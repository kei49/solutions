# Visualizing The Gender Gap In College Degrees

```python
import pandas as pd
import matplotlib.pyplot as plt
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']

cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_grey = (171/255, 171/255, 171/255)

fig = plt.figure(figsize=(18, 9))

def make_a_row(cats, num, ax_text, labelbottom=False):
    for sp in range(0, len(cats)):
        ax =plt.subplot(3, 6, num+sp+1)
        ax.plot(women_degrees['Year'], women_degrees[cats[sp]], c=cb_dark_blue, label='Women')
        ax.plot(women_degrees['Year'], 100-women_degrees[cats[sp]], c=cb_orange, label='Men')
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(cats[sp])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom=labelbottom)
        ax.set_yticks([0,100])
        ax.axhline(50, c=cb_light_grey, alpha=0.3)
        if sp==0:
            ax.text(ax_text[0][0], ax_text[0][1], ax_text[0][2])
            ax.text(ax_text[1][0], ax_text[1][1], ax_text[1][2])
        if sp==len(cats)-1:
            ax.text(ax_text[2][0], ax_text[2][1], ax_text[2][2])
            ax.text(ax_text[3][0], ax_text[3][1], ax_text[3][2])

ax_text1 = [[2002, 82, "Women"], [2005, 12, "Men"], [2005, 87, "Men"], [2001, 10, "Women"]]
ax_text2 = [[2002, 76, "Women"], [2005, 20, "Men"], [1970, 67, "Men"], [1970, 30, "Women"]]
ax_text3 = [[2002, 90, "Women"], [2005, 6, "Men"], [2005, 62, "Men"], [2001, 32, "Women"]]

make_a_row(stem_cats, 0, ax_text1)
make_a_row(lib_arts_cats, 6, ax_text2)
make_a_row(other_cats, 12, ax_text3, labelbottom=True)

plt.savefig('gender_degrees.png')
```
