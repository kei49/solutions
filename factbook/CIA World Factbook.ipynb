# Factbook

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('factbook.db')
q = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(q, conn)

q2 = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(q2, conn)
```

```python
q_min_population = "SELECT MIN(population) FROM facts"
pd.read_sql_query(q_min_population, conn)
```

```python
q_max_population = "SELECT MAX(population) FROM facts"
pd.read_sql_query(q_max_population, conn)
```

```python
q_min_population_growth = "SELECT MIN(population_growth) FROM facts"
pd.read_sql_query(q_min_population_growth, conn)
```

```python
q_max_population_growth = "SELECT MAX(population_growth) FROM facts"
pd.read_sql_query(q_max_population_growth, conn)
```

```python
q_no_population = "SELECT * FROM facts WHERE population = 0"
pd.read_sql_query(q_no_population, conn)
```

```python
q_max_population = "SELECT * FROM facts WHERE population = 7256490011"
pd.read_sql_query(q_max_population, conn)
```

```python
q_population = "SELECT population, population_growth, birth_rate, death_rate FROM facts WHERE population != 0 AND population !=7256490011"
popu = pd.read_sql_query(q_population, conn)
popu
```

```python
%matplotlib inline
popu['population'].plot.hist()
```

```python
popu['population_growth'].plot.hist()
```

```python
popu['birth_rate'].plot.hist()
```

```python
popu['death_rate'].plot.hist()
```

```python
q_population_density = "SELECT name, population, area_land FROM facts"
population_density = pd.read_sql_query(q_population_density, conn)
population_density['density'] = population_density['population'] / population_density['area_land']
population_density = population_density[population_density['density'].notnull()]
population_density.sort_values("density", ascending=False)
```

```python
population_density.sort_values("density", ascending=False)['density'][1:].plot.hist(bins=20)
```
