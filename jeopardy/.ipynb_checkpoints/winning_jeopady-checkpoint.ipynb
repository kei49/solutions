# Winning Jeoopady

In this project, we want to win on Jeopady, a popular TV show in the U.S.

To study the historical questions on Jeopady, we will analyze a dataset from [reddit](https://www.reddit.com/r/datasets/comments/1uyd0t/200000_jeopardy_questions_in_a_json_file/).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

jeopady = pd.read_csv('JEOPARDY_CSV.csv')
print(jeopady.shape)
jeopady.head()
```

```python
new_col = []
for col in jeopady.columns:
    col = col.replace(' ', '')
    new_col.append(col)
jeopady.columns = new_col
jeopady.columns
```

```python
jeopady[jeopady['Answer'].isnull()]
```

```python
jeopady = jeopady[jeopady['Answer'].notnull()]
```

```python
def normalize(string):
    try:
        string = string.lower().replace(',', '').replace(':', '').replace(';', '').replace('.', '').replace('\'', '').replace('\"','')
    except:
        print(string)
    return string
    
jeopady['clean_question'] = jeopady['Question'].apply(normalize)
jeopady['clean_answer'] = jeopady['Answer'].apply(normalize)
jeopady.head(20)
```
